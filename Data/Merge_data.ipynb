```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
sns.set()

import pylab
import scipy.stats
```

```python
btc = pd.read_csv("btc.csv")
btc["date"] = pd.to_datetime(btc["date"]).dt.strftime('%Y-%m-%d')
eth = pd.read_csv("eth.csv")
eth["date"] = pd.to_datetime(eth["date"]).dt.strftime('%Y-%m-%d')
btc.tail()
```

```python
btc_s = pd.read_csv("btc_sentiment.csv")
btc_s = btc_s.rename(columns={"polarity": "tw_polarity", "subjectivity": "tw_subjectivity"})
eth_s = pd.read_csv("eth_sentiment.csv")
eth_s = eth_s.rename(columns={"polarity": "tw_polarity", "subjectivity": "tw_subjectivity"})
btc_s.tail()
```

```python
btc_news = pd.read_csv("btc_news_sa.csv")
btc_news = btc_news.rename(columns={"Polarity": "news_polarity"})
eth_news = pd.read_csv("eth_news_sa.csv")
eth_news = eth_news.rename(columns={"Polarity": "news_polarity"})
btc_news.tail()
```

```python
btc_market = pd.read_csv("gemini_BTCUSD_day.csv")
btc_market["date"] = pd.to_datetime(btc_market["Date"]).dt.strftime('%Y-%m-%d')
btc_market = btc_market.drop(columns=['Date', 'Unix Timestamp', 'Symbol'])
eth_market = pd.read_csv("gemini_ETHUSD_day.csv")
eth_market["date"] = pd.to_datetime(eth_market["Date"]).dt.strftime('%Y-%m-%d')
eth_market = eth_market.drop(columns=['Date', 'Unix Timestamp', 'Symbol'])
btc_market.head()
```

```python
temp_btc = pd.merge(btc_s, btc_news, how='outer', on="date")
temp_btc = pd.merge(temp_btc, btc, how='outer', on="date")
temp_btc = pd.merge(temp_btc, btc_market, how='outer', on="date")
temp_btc.tail()
```

```python
temp_eth = pd.merge(eth_s, eth_news, how='outer', on="date")
temp_eth = pd.merge(temp_eth, eth, how='outer', on="date")
temp_eth = pd.merge(temp_eth, eth_market, how='outer', on="date")
temp_eth.tail()
```

```python
temp_btc.to_csv('merged_btc.csv')
temp_eth.to_csv('merged_eth.csv')
```
